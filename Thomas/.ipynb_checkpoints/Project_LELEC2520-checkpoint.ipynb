<h2> <center> Part 6 : Further analysis </center> </h2>

In this part, the aim is to optimize some aspects of the grid. Until here, fixed values of active and reactive powers were given to the generators and the PandaPower tool computed the Load Flow. But there might be some situations where the grid operator wants to minimize the power system's cost. 

To do so, a cost function has to be introduced in order to know the quantity to minimize inside the Power System.

However, the minimization (or the maximization) cannot be done under any conditions. Indeed, if a maximization of a power through a line A is focused, a strong care has to be made on the overloading of the transformers inside the power system. What is the maximum permissible loading of the transformers in order to keep their lifespan as long as possible ? In a word, the optimization comes with constraints to keep the power system viable.

$$ \textbf{min (or max)} \; Cost\_Function$$

$$
\begin{cases} 
Constraint_1 \\
Constraint_2 \\
 . . . \\
Constraint_n 
\end{cases}
$$
